Classes: The template of Objects

```python
class Info:
  def __init__(self,Name,Address,salary):
      self.Name = Name
      self.Address = Address
      self.salary = salary
  def display_format(self):
    return print(f"Name: {self.Name}\nAddress: {self.Address}\nsalary:{self.salary:,.0f}")
```

```python
add1 = Info(Name,Address,salary) # object
```

```python
add1.salary # attribute
```

```python
add1.display_format() #method
```
