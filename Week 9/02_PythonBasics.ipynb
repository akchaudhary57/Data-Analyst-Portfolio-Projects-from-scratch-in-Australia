Loops
* for loops
* While loops

```python
numbers = [0,1,2,3,4,5]

for number in numbers:
  print(number)
```

```python
year_Exp = {
    'luke':3,'kelly':6,'john':9,'adam':10
}
```

```python
for key,value in year_Exp.items():
  if value>5:
    print(key,'has',value,'year of experience')
```

```python
while True:
  print("Data Nerd")
```

```python
count = 1
while count<5:
  count = count+1
  print(count)
```
